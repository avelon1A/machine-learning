```python
import numpy as np 
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,5,9])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,5,9])

plt.bar(x,y,width=0.1)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,5,9])

plt.barh(x,y,height=0.5)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,5,9])

plt.plot(x,y)
plt.xlabel("number")
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,5,9])

plt.plot(x,y,marker ='o')
```

```python
#subplot--->
#if you want to create more than one graph 
#to compare
x = np.array([25,45,78,90])
y = np.array([67,89,23,45])
plt.subplot(1,2,1)
plt.plot(x,y,color='red')

a = np.array([25,45,78,90])
b = np.array([67,89,23,45])
plt.subplot(1,2,2)
plt.plot(a,b,'--')
```
